```python
from scipy import misc

import matplotlib.pyplot as plt
import numpy as np

```

```python
#----------------------------------------------------------------------------------------#
# Function Definition

def E(x):
    return (x-1)*(x-2)*(x-3)*(x-5) # =  x**4 - 11 x**3 + 41 x**2 - 61 x + 30
#----------------------------------------------------------------------------------------#
# Plot function

x = np.arange(0, 6.0, 0.01)
y = E(x)
plt.figure(figsize=(10,10))
plt.plot(x, y,'yellow')
#plt.show()

#----------------------------------------------------------------------------------------#
# Descent de Gradient.

step = 0.001 # learning rate
nb_max_iter = 1000 # Nb max d'iteration
eps = 0.01 # stop condition

x0 = 5 # start point
y0 = E(x0)
plt.scatter(x0, E(x0))

cond = eps + 10.0 # start with cond greater than eps (assumption)
nb_iter = 0 
tmp_y = y0
while cond > eps and nb_iter < nb_max_iter:
    x0 = x0 - step * misc.derivative(E, x0)
    y0 = E(x0)
    nb_iter = nb_iter + 1
    cond = abs( tmp_y - y0 )
    tmp_y = y0
    
    plt.scatter(x0, y0, c="b")
print("****_****"*5,"\n")
print("Le Xmin : "+"** **",x0,"E(Xmin) : "+"** **",y0,"Le nombre d'itération"+"** **",nb_iter)
print("****_****"*5,"\n")
print("epsilon = ",eps," nbre max itération : ",nb_max_iter)
print("****_****"*5,"\n")
plt.title("La Descent de Gradient à epsilon ="+str(eps)+" et max_iteration ="+str(nb_max_iter))
plt.grid()

plt.savefig("gradient_descent_1d_python.png", bbox_inches='tight')
plt.show()
```
