# Prétraitement et visualisation de données

# A. Normalisation des données 

```python
# Imports 
import numpy as np
from sklearn.preprocessing import *
```

```python
#1 Créer a matrice 
X = np.array([[1,-1,2],[2,0,0], [0,1,-1]])
#Visualisation de la matrice , moyenne , et variance 
print("*** ***"*5,"\n")
print("La matrice X : \n")
print(X)
print("*** ***"*5,"\n")
print(" la moyenne de X : ", np.mean(X))
print("*** ***"*5,"\n")
print("la varaince de X : " ,np.var(X))
print("*** ***"*5,"\n")
```

```python
#Normalisation de X
print("*** ***"*5,"\n")
Xnorm = scale(X)
print("*** ***"*5,"\n")
print("La Matrice X après normalisation :\n")
print("*** ***"*5,"\n")
print(Xnorm)
print("*** ***"*5,"\n")
print(" la moyenne de X : ", np.mean(Xnorm))
print("*** ***"*5,"\n")
print("la varaince de X : " ,np.var(Xnorm))
print("*** ***"*5,"\n")

"""
Apres normalisation de la matice X la moyen et la variance n'a pas changé
tandis que sa représentation l'est
"""
```

# B. NormalisationMinMax


```python
X2 = np.array([[1,-1,2],[2,0,0], [0,1,-1]])
#Visualisation de la matrice , moyenne , et variance 
print("La matrice X2 : \n")
print(X2)
print("*** ***"*5,"\n")
print("la moyenne de X2 : ", np.mean(X2))
print("*** ***"*5,"\n")
print("la varaince de X2 : " ,np.var(X2))
print("*** ***"*5,"\n")

mmscaler = MinMaxScaler(feature_range=(0,1))
X2MinMaxnorm = mmscaler.fit_transform(X2)
print("*** ***"*5,"\n")
print("La Matrice X2 après MinMax normalisation :\n")
print(X2MinMaxnorm)
print("*** ***"*5,"\n")
print(" la moyenne de X2MinMaxnorm : ", np.mean(X2MinMaxnorm))
print("*** ***"*5,"\n")
print("la varaince de X2MinMaxnorm : " ,np.var(X2MinMaxnorm))
print("*** ***"*5,"\n")
```

# C. Visualisation des données

```python
from sklearn import * 
import matplotlib.pyplot as plt
```

```python
# charger données
iris = datasets.load_iris() 

plt.figure(figsize=(20,7))
plt.subplot(161)
plt.scatter(iris.data[:,0], iris.data[:, 1], c=iris.target)
plt.title("0 vs 1")
plt.subplot(162)
plt.scatter(iris.data[:,0], iris.data[:, 2], c=iris.target)
plt.title("0 vs 2")
plt.subplot(163)
plt.scatter(iris.data[:,0], iris.data[:, 3], c=iris.target)
plt.title("0 vs 3")
plt.subplot(164)
plt.scatter(iris.data[:,1], iris.data[:, 2], c=iris.target)
plt.title("1 vs 2")
plt.subplot(165)
plt.scatter(iris.data[:,1], iris.data[:, 3], c=iris.target)
plt.title("1 vs 3")
plt.subplot(166)
plt.scatter(iris.data[:,2], iris.data[:, 3], c=iris.target)
plt.title("2 vs 3")
plt.show()

#
#""""
#La meilleure visualisation c'est 2vs3 car les données sont indépendablement séparable.
#""""
```

# D . Réduction de Dimension et visualisation de données 

ACP & LDA

```python
from sklearn.decomposition import PCA 
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
Iris_pca = PCA(n_components=2).fit_transform(iris.data)
#Iris_pca =PCA(n_componenets=2).fit(iris.data).transform(iris.data)
Iris_lda = LDA(n_components=2).fit(iris.data,iris.target).transform(iris.data)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(121)
plt.scatter(Iris_pca[:, 0], Iris_pca[:, 1], c=iris.target)
plt.title("Iris PCA")
plt.subplot(122)
plt.scatter(Iris_lda[:, 0], Iris_lda[:, 1], c=iris.target)
plt.title("Iris LDA")
plt.show()
```
